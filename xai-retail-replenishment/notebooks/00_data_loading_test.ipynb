# Data Loading Smoke Test

This notebook verifies that `src/data/load_data.py` correctly loads Rossmann and M5 raw files.

Run cells from top to bottom.

```python
from pathlib import Path
import sys
import importlib

# Resolve project root whether notebook cwd is project root or notebooks/
cwd = Path.cwd().resolve()
project_root = cwd if cwd.name == "xai-retail-replenishment" else cwd.parent

src_path = project_root / "src"
if str(src_path) in sys.path:
    sys.path.remove(str(src_path))
sys.path.insert(0, str(src_path))

# Always reload local module so notebook picks latest code edits
import data.load_data as load_data_module
importlib.reload(load_data_module)

load_m5 = load_data_module.load_m5
load_rossmann = load_data_module.load_rossmann
load_dataset = load_data_module.load_dataset

print("Project root:", project_root)
print("Source path prioritized:", src_path)
print("Loaded module file:", load_data_module.__file__)
```

```python
rossmann_dir = project_root / "data" / "raw" / "rossmann"
print("Rossmann directory:", rossmann_dir)

ross = load_rossmann(str(rossmann_dir))
print("Rossmann keys:", list(ross.keys()))
for k, v in ross.items():
    print(f"{k}: shape={v.shape}")
```

```python
m5_dir = project_root / "data" / "raw" / "m5"
print("M5 directory:", m5_dir)

try:
    m5 = load_m5(str(m5_dir))
    print("M5 keys:", list(m5.keys()))
    for k, v in m5.items():
        print(f"{k}: shape={v.shape}")
except FileNotFoundError as e:
    print("M5 not ready yet:", e)
    m5 = None
```

```python
print("\nRossmann preview:")
print(ross["train"].head(2))
print("\nRossmann train dtypes:")
print(ross["train"].dtypes.head(10))

if m5 is not None:
    print("\nM5 calendar preview:")
    print(m5["calendar"].head(2))
```

```python
# Dispatcher smoke test
r2 = load_dataset("rossmann", str(rossmann_dir))
print("Dispatcher rossmann keys:", list(r2.keys()))

if m5 is not None:
    m2 = load_dataset("m5", str(m5_dir))
    print("Dispatcher m5 keys:", list(m2.keys()))
```
